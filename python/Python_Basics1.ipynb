*   NO CURLY BRACES. **INDENTATION MATTERS.**




```python
a, b = 12, 33
if (a < b):
      print('hello')
      if a < 10:
        print( a, ' is less than 10')
```

```python
# a = 2
# b = 5
a, b = 20, 5
if a > b:
    print('a is greater than b') # yes
    if a < 10:
        print('a is less than 10') # no
    elif a < 100: # else if
        print('a is greater than 9 and less than 100') # yes
else:
    print('a is not greater than b') # no
```

## While loop

```python
n = 0
while True:
    print(n)
    # n++, n--, --n, ++n is not supported
    n += 1
    n = n+1
    n = n-1
    n -= 1
    n *= 2
    n /= 2
    n += 1
    if n == 5: break
```

```python
i = 1
while i < 6:
    print(i)
    i += 2
```

```python
i = 1
while i < 100:
    if i % 10 == 0:
        print(i)
    elif i % 23 == 0:
        print(i, 'need to break')
        break
    i += 3
```

## Slicing

```python
temp = arr[::-1]
print(temp)
```

## copy a list

```python
foo = [1, 2, 3]
bar = foo
bar.append(4)
print(foo, bar, '\n')

```

```python
foo = [1, 2, 3]
bar = foo.copy()
bar.append(5)
print(foo, bar, '\n')
```

```python
foo = [1, 2, []]
bar = foo
bar[-1].append(4)
print(foo, bar, '\n')

foo = [1, 2, []]
bar = foo.copy()
bar[-1].append(5)
print(foo, bar, '\n')

```

```python
import copy
foo = [1, 2, []]
bar = copy.deepcopy(foo)
bar[-1].append(5)
print(foo, bar, '\n')
```
