```python
## Way to print variables

print(a, 'apples')
# formatted print; similar to printf(f"{a} apples");
print(f"{a} apples")

print(f"{a} apples, {b} oranges") # 5 apples, 10 oranges
print("%2d apples, %2d oranges" % (a, b))

print(a, 'apples', b, 'oranges')
```

```python
a = 3.5
a = a + 1 # 4.5
a = a - 2 # 2.5
a = a * 2 # 5
a = a / 2 # 2.5
a = a ** 2 # 2.5^2=6.25
a = a % 5 # 1.25 # modulus
print(a)
```

```python
# Power

b = 3 ** 5 # 3^5
print(b)

b = 6 % 4 # 2
print(b)
```
